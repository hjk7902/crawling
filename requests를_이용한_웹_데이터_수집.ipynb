```python
from requests.adapters import HTTPAdapter
```

```python
from requests.exceptions import ConnectionError
```

```python
github_adapter = HTTPAdapter(max_retries=3)
session = requests.Session()
```

```python
session.mount("https://api.github.com", github_adapter)
```

```python
try:
    session.get("https://api.github.com")
except ConnectionError as ce:
    print(ce)
```

```python
#  
import requests
from bs4 import BeautifulSoup

url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn'
movie_ranking = requests.get(url)
print(movie_ranking)

soup = BeautifulSoup(movie_ranking.content, 'html.parser')
title_list = soup.select('div.tit3 > a')
print(title_list[0])

for i, title in enumerate(title_list):
    print(f'{i+1}: {title.text}')
```

```python
import pandas as pd
movie_df = pd.DataFrame([], columns=['rank', 'title'])

for i, title in enumerate(title_list):
    movie_df = movie_df.append({'rank': i+1, 'title': title.text}, ignore_index=True)
```

```python
movie_df.to_csv("movie_rank.csv", index=False, encoding='utf-8-sig')
```
